```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
df = pd.read_excel('Parts Data.xlsx')
```

```python
df = df[~df['VendorPartNo'].isna()]
df = df[~df['Category'].isna()]
df
```

```python
df['IsDevelopment']=df['IsDevelopment'].apply(lambda x: 1 if x =='Yes' else 0)
df['Worldpac']=df['Worldpac'].apply(lambda x: 1 if x =='Yes' else 0)
df['PartsAuthority']=df['PartsAuthority'].apply(lambda x: 1 if x =='Yes' else 0)
df['SSF']=df['SSF'].apply(lambda x: 1 if x =='Yes' else 0)
df['Total VIO']=df.apply(lambda x: x['VIO_US'] + x['VIO_CANADA'], axis=1)
df
```

```python
df.columns
```

```python
melt_df=df.melt(id_vars=['VendorPartNo', 'Make', 'vehicleMaxQty', 'Category', 'IsDevelopment',
       'Worldpac', 'PartsAuthority', 'SSF', 'Total VIO',
       'FirstSoldDate'])
melt_df
```

```python
def year_extract(x):
    if '2016' in x:
        return '2016'
    elif '2017' in x:
        return '2017'
    elif '2018' in x:
        return '2018'
    elif '2019' in x:
        return '2019'
    elif '2020' in x:
        return '2020'
    elif x=='UnitsSold360':
        return '2019'
    else:
        return np.nan
```

```python
melt_df['Sales Year'] = melt_df.apply(lambda x: year_extract(x['variable']), axis=1)
melt_df
```

```python
melt_df['Sales By'] = melt_df.apply(lambda x: x['variable'].split('(')[0] + ' Sales count', axis=1)
melt_df
```

```python
melt_df.columns
```

```python
melt_df = melt_df[['VendorPartNo', 'Make', 'vehicleMaxQty', 'Category', 'IsDevelopment',
       'Worldpac', 'PartsAuthority', 'SSF', 'Total VIO',
       'FirstSoldDate','value', 'Sales Year', 'Sales By']]
melt_df
```

```python
x = melt_df.pivot_table(index=['VendorPartNo','Sales Year'], columns='Sales By', values='value', dropna=False)
x
```

```python
t = x.reset_index()
t
```

```python
t = t.fillna({'UnitsSold360 Sales count':0})
```

```python
f=f[['VendorPartNo', 'Sales Year', 'Make', 'Category', 'IsDevelopment', 'Total VIO', 'FirstSoldDate', 
     'vehicleMaxQty', 'PartsAuthority', 'SSF', 'Worldpac', 'PartsAuthority Sales count', 'SSF Sales count', 
     'WorldPac Sales count', 'UnitsSold360 Sales count', 'Total Units Sold Sales count']].copy()
```

```python
def find_year_rank(x):
    v = int(x['Sales Year']) - x['FirstSoldDate'].year
    return v + 1 
```

```python
f['SalesYearRank'] = f.apply(lambda x: find_year_rank(x), axis=1)
f
```
