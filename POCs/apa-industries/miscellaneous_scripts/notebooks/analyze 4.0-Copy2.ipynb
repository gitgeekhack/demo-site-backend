```python
from scipy.stats import chi2_contingency
from scipy.stats import chi2
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from statsmodels.stats.outliers_influence import variance_inflation_factor
```

```python
df = pd.read_csv('transformed_data.csv')
```

```python
def calc_vif(X):

    # Calculating VIF
    vif = pd.DataFrame()
    vif["variables"] = X.columns
    vif["VIF"] = [variance_inflation_factor(X.values, i) for i in range(X.shape[1])]

    return(vif)
```

```python
x = df[['Sales Year', 'Make', 'Category', 'IsDevelopment', 'Total VIO',
       'vehicleMaxQty', 'PartsAuthority', 'SSF',
       'Worldpac', 'PartsAuthority Sales count', 'SSF Sales count',
       'WorldPac Sales count', 'UnitsSold360 Sales count',
       'Total Units Sold Sales count', 'SalesYearRank']].apply(lambda x: x.factorize()[0])
X = x.iloc[:,:-1]
calc_vif(X)
```
