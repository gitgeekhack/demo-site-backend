```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
df = pd.read_excel('Parts Data.xlsx')
```

```python
df.head(10)
```

```python
df.columns
```

```python
melt_df=df.melt(id_vars=['VendorPartNo', 'Make', 'vehicleMaxQty', 'Category', 'IsDevelopment',
       'Worldpac', 'PartsAuthority', 'SSF', 'VIO_US', 'VIO_CANADA',
       'FirstSoldDate'], var_name='Data Provider')
```

```python
melt_df.rename(columns={'value':'Sales Count'}, inplace=True)
```

```python
melt_df = melt_df[melt_df['Sales Count']>0]
```

```python
# x = melt_df[melt_df['VendorPartNo']=='C031747'].copy()
x = melt_df.copy()
```

```python
x[x['VendorPartNo']=='B-6973']
```

```python
def year_extract(x):
    if '2017' in x:
        return '2017'
    elif '2018' in x:
        return '2018'
    elif '2019' in x:
        return '2019'
    elif '2020' in x:
        return '2020'
    elif x=='UnitsSold360':
        return '2020'
    else:
        return np.nan
```

```python
x['Year'] = x.apply(lambda x: year_extract(x['Data Provider']), axis=1)
x['IsDevelopment'] = x.apply(lambda x:1 if not pd.isnull(x['IsDevelopment']) else 0 , axis=1)
```

```python
# pd.pivot_table(x, values=['Sales Count'], index=['VendorPartNo', 'Make','Category','Year'],
#                     aggfunc={'Sales Count': np.mean})
data2 = pd.pivot_table(x, values=['Sales Count'], index=['VendorPartNo', 'VIO_US', 'VIO_CANADA','Year'],
                    aggfunc={'Sales Count': np.mean}).unstack()
data = pd.pivot_table(x, values=['Sales Count'], index=['VendorPartNo', 'Year'],
                    aggfunc={'Sales Count': np.mean}).unstack()

data2
```

```python
# x.groupby(['VendorPartNo', 'Year'])['Sales Count'].agg('mean').unstack()
data2 = data2.reset_index()
data2
```

```python
fig = data[:10].plot(kind='bar', figsize=(50, 20), fontsize=16).get_figure()
fig.savefig('test.pdf', bbox_inches='tight')
```

```python
data3 = pd.pivot_table(x, values=['Sales Count'], index=['VendorPartNo', 'Make', 'vehicleMaxQty', 'Category', 'VIO_US', 'VIO_CANADA', 'IsDevelopment','Year'],
                    aggfunc={'Sales Count': np.mean}).unstack()
data3
```

```python
data3 = data3.reset_index()
```

```python
t = data3.corr(method ='pearson')
f = plt.figure(figsize=(15, 10))
plt.matshow(t, fignum=f.number)
plt.xticks(range(t.shape[1]), t.columns, fontsize=14, rotation=45)
plt.yticks(range(t.shape[1]), t.columns, fontsize=14)
cb = plt.colorbar()
cb.ax.tick_params(labelsize=14)
plt.title('Correlation Matrix', fontsize=16)
```

```python
t = data3.corr(method ='kendall')
f = plt.figure(figsize=(15, 10))
plt.matshow(t, fignum=f.number)
plt.xticks(range(t.shape[1]), t.columns, fontsize=14, rotation=45)
plt.yticks(range(t.shape[1]), t.columns, fontsize=14)
cb = plt.colorbar()
cb.ax.tick_params(labelsize=14)
plt.title('Correlation Matrix', fontsize=16)
```

```python
t = data3.corr(method ='spearman')
f = plt.figure(figsize=(15, 10))
plt.matshow(t, fignum=f.number)
plt.xticks(range(t.shape[1]), t.columns, fontsize=14, rotation=45)
plt.yticks(range(t.shape[1]), t.columns, fontsize=14)
cb = plt.colorbar()
cb.ax.tick_params(labelsize=14)
plt.title('Correlation Matrix', fontsize=16)
```
