```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('transformed_data.csv')
```

```python
df
```

```python
df.columns
```

```python
t = df[['Sales Year', 'Make', 'Category', 'IsDevelopment', 'Total VIO',
       'vehicleMaxQty', 'PartsAuthority', 'SSF',
       'Worldpac', 'PartsAuthority Sales count', 'SSF Sales count',
       'WorldPac Sales count', 'UnitsSold360 Sales count',
       'Total Units Sold Sales count', 'SalesYearRank']].apply(lambda x: x.factorize()[0]).corr(method='pearson').round(2)
t
```

```python
f = plt.figure(figsize=(15, 8))
plt.matshow(t, fignum=f.number)
plt.xticks(range(t.shape[1]), t.columns, fontsize=14, rotation=90)
plt.yticks(range(t.shape[1]), t.columns, fontsize=14)
cb = plt.colorbar()
cb.ax.tick_params(labelsize=10)
plt.title('Correlation Matrix', fontsize=10)
```

```python
t.to_csv('Correlations pearson.csv')
```
